```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries loaded successfully")
```

 * **Data Loading**

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)

df.head()
df.shape
```

```python
df.info()
```

*  **Data Understanding**

```python
df.describe()
```

 * **Visualizationisualization**

```python
plt.figure(figsize=(6,4))
sns.countplot(x='sex', data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['total_bill'], kde=True)
plt.title("Total Bill Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.title("Total Bill vs Tip")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='day', data=df)
plt.title("Sales Count by Day")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='sex', y='total_bill', data=df)
plt.title("Average Bill by Gender")
plt.show()
```

 * **Conclusion:**

  This project demonstrates how exploratory data analysis helps understand customer spending and tipping behavior using visualization techniques.

 * **Insights:**
1. Most customers fall within a certain bill range.
2. Higher total bills generally lead to higher tips.
3. Some days have more customer activity than others.
4. Spending pattern varies slightly by gender.
